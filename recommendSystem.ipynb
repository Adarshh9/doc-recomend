```python
x = [] 
z = []
for i in df["specilizationDesciption"]:
    x.append(i.split(", "))
    
for i in df["userProblems"]:
    z.append(i.split(", "))
```

```python
df["specilizationDesciption"] = df["specilizationDesciption"].apply(lambda x : [i.replace(" ","") for i in x])
df["specialist"] = df["specialist"].apply(lambda x : [i.replace(" ","") for i in x])
df["location"] = df["location"].apply(lambda x : [i.replace(" ","") for i in x])
df["userProblems"] = df["userProblems"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
df2 = df[["doctorName","text","patientsVisited"]]
```

```python
df2["text"] = df2["text"].apply(lambda x : [i.lower() for i in x])
```

```python
df2.head()
```

```python
df2["text"] = df2["text"].apply(lambda x: " ".join(x))
```

Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words="english")
```

```python
vectors = cv.fit_transform(df2["text"]).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])) , reverse=True , key=lambda x:x[1])[1:6]
```

```python
user_location = "borivali"
user_problem = "stomach pain"
combined_input = f"{user_location} {user_problem}"
```

```python
input_vector = cv.transform([combined_input])

similarity_score = cosine_similarity(input_vector,vectors)

top_doctors_indices = similarity_score.argsort()[0][::-1]

recommended_doctors = df.loc[top_doctors_indices, ['doctorName', 'location', 'specialist' , 'specilizationDesciption' , 'patientsVisited']][:5]

print(recommended_doctors)
```
